```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#ploting a line graph
x=[0,1,20,3,40,5]
y=[0,5,10,15,20,25]

#adding labels and adding font
plt.xlabel('x-axis ',fontdict={'fontname':'monospace','fontsize':10})
plt.ylabel('y-axis')

#adding title
plt.title('Simple Graph',color='r')

#plotting
plt.plot(x,y)
plt.show()
```

```python
#adding style to line
plt.plot(x,y,ls='dotted',color='m')#linestyle can dashed--,dtted:,solid-,dashdot-.

#removing decimals
plt.yticks([5,10,15,20,25,30,35,40])
plt.show()
```

```python
#ploting a bar graph
import matplotlib.pyplot as plt
x=['monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
y=[100,90,88,54,67,56,70]
plt.title('Sales of company')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.bar(x,y,color=['r','y','b','orange','k','grey','blue'])
plt.pause(0.001)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#adding Nessasary Files and libraryes
gas =pd.read_csv('gas_prices.csv')
gas
```

```python
#ploting line graph for different for comparison
plt.title('Gas price comparison')
plt.xlabel('Year')
plt.ylabel('Price (us dollar)')


plt.plot(gas.Year,gas.USA,label='Usa',marker='+')
plt.plot(gas.Year,gas.Canada,label='Canada',marker='+')
plt.plot(gas.Year,gas['South Korea'],label='South Korea',marker='+')
plt.plot(gas.Year,gas.Australia,label='Australia',marker='^')

plt.legend()

plt.show()
```

```python
# using fifa data
fifa = pd.read_csv('fifa_data.csv')
fifa
```

```python
plt.xlabel('slikk level')
plt.ylabel('no. of players')
plt.hist(fifa.Overall,bins=[40,50,60,70,80,90,100],color='y')
plt.xticks()
plt.show()
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
left,right
```

```python
#pie chart for comparison of left vs right foot player
plt.pie(  [left,right],labels=['Left','Right'],autopct='%.2f %%', explode=[0.1,0])
plt.show()
```

```python
plt.figure(figsize=(10,10))
realmadrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
barcalone=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
revs=fifa.loc[fifa.Club=='New England Revolution']['Overall']
juventus=fifa.loc[fifa.Club=='Juventus']['Overall']
#plotting a boox plot
boxes = plt.boxplot([barcalone,realmadrid,revs,juventus] , patch_artist=True)
#style individula boxes
for box in boxes['boxes']:
    box.set(color='c',linewidth=2)
    box.set(facecolor='k')




plt.title('Club Comparison')
plt.xlabel('Club')
plt.ylabel('Overall Rating')
plt.xticks([1,2,3,4],['FC Barcelona','Real Madrid','New England','juventus'])

plt.show()
```
